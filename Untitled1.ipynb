```python
# Sorting dataframe by rating so we can remove some of 5 and 4 star ratings 
# Doing that so we have balances dataset
df['rating'] = df['rating'].str[:1] # rating is in firs character, usaly looks -> 2,0 su 5 stelle
df = df.sort_values(by=['rating'], ascending=False)
df = df.reset_index(drop=True)

df_ratings5 = df.loc[:5414]
df_ratings4 = df.loc[5415:6373]
df_other_ratings = df.loc[6374:]
```

```python
# Selecting 5 star ratings which have between 25 and 75 characters 
mask = (df_ratings5['content'].str.len() >= 5) & (df_ratings5['content'].str.len() <= 150)
df_ratings5 = df_ratings5.loc[mask]
# Taking every 5th rating
df_ratings5 = df_ratings5.sample(frac=1).reset_index(drop=True)
df_ratings5 = df_ratings5.iloc[::20, :]
```

```python
# Selecting 4 star rating which have between 25 and 75 characters 
mask = (df_ratings4['content'].str.len() >= 5) & (df_ratings4['content'].str.len() <= 30)
df_ratings4 = df_ratings4.loc[mask]
```

```python
# Merging frames together
frames = [df_ratings5, df_ratings4, df_other_ratings]
df = pd.concat(frames)
df = df.reset_index(drop=True)
```

```python
## NLTK Preprocessing
# Preparing our content column for nltk processing
# all letters to lower andd convert to string type
df = df.loc[:910]
df['content'] = df['content'].str.lower().astype(str)
df['rating'] = df['rating'].astype(int)

df['rating'] = df['rating'].replace(1, 'negative') 
df['rating'] = df['rating'].replace(2, 'negative') 
df['rating'] = df['rating'].replace(3, 'neutral') 
df['rating'] = df['rating'].replace(4, 'positive') 
df['rating'] = df['rating'].replace(5, 'positive') 

#df['rating'] = df['rating'].astype(str)
```

```python
## Tokenize words in content and content_no_stopwords columns
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
df["content_tokens"] = df["content"].apply(tokenizer.tokenize)
```

```python
stop = nltk.corpus.stopwords.words('italian')
df['content_noSW_tokens'] = df['content_tokens'].apply(lambda x: [item for item in x if item not in stop])
```

```python
## Prepare dataset, split to train and test
#First shufle data, becouse its sorted by rating and reset index
from sklearn.model_selection import train_test_split

df = df.sample(frac=1).reset_index(drop=True) # shuffle data

list_corpus = df["content_tokens"].tolist()
list_corpus = [' '.join(review) for review in list_corpus]
list_labels = df["rating"].tolist()

X_train, X_test, y_train, y_test = train_test_split(list_corpus, list_labels, test_size=0.20, 
                                                                                random_state=40)
```

```python
## Function to vectorize data
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

def cv(data):
    count_vectorizer = CountVectorizer()
    emb = count_vectorizer.fit_transform(data)
    return emb, count_vectorizer

X_train_counts, count_vectorizer = cv(X_train)
X_test_counts = count_vectorizer.transform(X_test)
```

```python
from sklearn.decomposition import PCA, TruncatedSVD
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

def plot_LSA(test_data, test_labels, savepath="PCA_demo.csv", plot=True):
        lsa = TruncatedSVD(n_components=2)
        lsa.fit(test_data)
        
        lsa_scores = lsa.transform(test_data)
        print(len(lsa_scores))
        color_mapper = {label:idx for idx,label in enumerate(set(test_labels))}
        color_column = [color_mapper[label] for label in test_labels]
        print(len(color_column))
        colors = ['cyan','yellow','magenta']
        if plot:
            plt.scatter(lsa_scores[:,0], lsa_scores[:,1], s=8, alpha=.8, c=color_column)#, cmap=matplotlib.colors.ListedColormap(colors)
            red_patch = mpatches.Patch(color='cyan', label='negative')
            green_patch = mpatches.Patch(color='yellow', label='neutral')
            blue_patch = mpatches.Patch(color='magenta', label='positive')
            
            
            plt.legend(handles=[blue_patch, green_patch, red_patch], prop={'size': 15})


fig = plt.figure(figsize=(10, 10))          
plot_LSA(X_train_counts, y_train)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(C=30.0, class_weight='balanced', solver='newton-cg', 
                         multi_class='multinomial', n_jobs=-1, random_state=40)
clf.fit(X_train_counts, y_train)

y_predicted_counts = clf.predict(X_test_counts)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report

def get_metrics(y_test, y_predicted):  
    # true positives / (true positives+false positives)
    precision = precision_score(y_test, y_predicted, pos_label=None,
                                    average='weighted')             
    # true positives / (true positives + false negatives)
    recall = recall_score(y_test, y_predicted, pos_label=None,
                              average='weighted')
    
    # harmonic mean of precision and recall
    f1 = f1_score(y_test, y_predicted, pos_label=None, average='weighted')
    
    # true positives + true negatives/ total
    accuracy = accuracy_score(y_test, y_predicted)
    return accuracy, precision, recall, f1

accuracy, precision, recall, f1 = get_metrics(y_test, y_predicted_counts)
print("accuracy = %.3f, precision = %.3f, recall = %.3f, f1 = %.3f" % (accuracy, precision, recall, f1))
```

```python
import numpy as np
import itertools
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.winter):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=30)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=20)
    plt.yticks(tick_marks, classes, fontsize=20)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", 
                 color="white" if cm[i, j] < thresh else "black", fontsize=40)
    
    plt.tight_layout()
    plt.ylabel('True label', fontsize=30)
    plt.xlabel('Predicted label', fontsize=30)

    return plt

```

```python
cm = confusion_matrix(y_test, y_predicted_counts)
fig = plt.figure(figsize=(10, 10))
plot = plot_confusion_matrix(cm, classes=['Negative','Neutral','Positive'], normalize=False, title='Confusion matrix')
plt.show()
print(cm)
```

```python
def get_most_important_features(vectorizer, model, n=5):
    index_to_word = {v:k for k,v in vectorizer.vocabulary_.items()}
    
    # loop for each class
    classes ={}
    for class_index in range(model.coef_.shape[0]):
        word_importances = [(el, index_to_word[i]) for i,el in enumerate(model.coef_[class_index])]
        sorted_coeff = sorted(word_importances, key = lambda x : x[0], reverse=True)
        tops = sorted(sorted_coeff[:n], key = lambda x : x[0])
        bottom = sorted_coeff[-n:]
        classes[class_index] = {
            'tops':tops,
            'bottom':bottom
        }
    return classes

importance = get_most_important_features(count_vectorizer, clf, 10)
"""
print(importance[0])
print('\n')
print(importance[1])
print('\n')
print(importance[2])
print('\n')
"""
```

```python
def plot_important_words(top_scores, top_words, neut_scores, neut_words, bottom_scores, bottom_words, name):
    y_pos = np.arange(len(top_words))
    
    top_pairs = [(a,b) for a,b in zip(top_words, top_scores)]
    top_pairs = sorted(top_pairs, key=lambda x: x[1])
    
    neut_pairs = [(a,b) for a,b in zip(neut_words, neut_scores)]
    neut_pairs = sorted(neut_pairs, key=lambda x: x[1])
    
    bottom_pairs = [(a,b) for a,b in zip(bottom_words, bottom_scores)]
    bottom_pairs = sorted(bottom_pairs, key=lambda x: x[1], reverse=True)
    
    top_words = [a[0] for a in top_pairs]
    top_scores = [a[1] for a in top_pairs]

    neut_words = [a[0] for a in neut_pairs]
    neut_scores = [a[1] for a in neut_pairs]
    
    bottom_words = [a[0] for a in bottom_pairs]
    bottom_scores = [a[1] for a in bottom_pairs]
    
    fig = plt.figure(figsize=(14, 6))  
    

    plt.subplot(131)
    plt.barh(y_pos,bottom_scores, align='center', alpha=0.5)
    plt.title('Positive', fontsize=20)
    plt.yticks(y_pos, bottom_words, fontsize=14)
    plt.suptitle('Key words', fontsize=16)
    plt.xlabel('Importance', fontsize=20)
    
    plt.subplot(132)
    plt.barh(y_pos,neut_scores, align='center', alpha=0.5)
    plt.title('Neutral', fontsize=20)
    plt.yticks(y_pos, neut_words, fontsize=14)
    plt.suptitle(name, fontsize=16)
    plt.xlabel('Importance', fontsize=20)
    
    plt.subplot(133)
    plt.barh(y_pos,top_scores, align='center', alpha=0.5)
    plt.title('Negative', fontsize=20)
    plt.yticks(y_pos, top_words, fontsize=14)
    plt.suptitle(name, fontsize=16)
    plt.xlabel('Importance', fontsize=20)
    
    plt.subplots_adjust(wspace=2)

    plt.show()

top_scores = [a[0] for a in importance[0]['tops']]
top_words = [a[1] for a in importance[0]['tops']]
neut_scores = [a[0] for a in importance[1]['tops']]
neut_words = [a[1] for a in importance[1]['tops']]
bottom_scores = [a[0] for a in importance[2]['tops']]
bottom_words = [a[1] for a in importance[2]['tops']]

plot_important_words(top_scores, top_words, neut_scores, neut_words, bottom_scores, bottom_words, "Most important words for relevance BOW")
```
